```python
import pandas as pd
```

```python
# Load in your csv.gz's of results for each year extracted.
df2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz') 
df2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
# Concatenate the data into 1 dataframe for the remainder of the analysis.
df = pd.concat([df2000, df2001])
display(df.head(),df.tail())
```

```python
df.info()
```

# How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
# either nonzero budget or revenue
valid_finance = df[(df['budget']>0) | (df['revenue']>0)]
valid_finance.head()
```

```python
valid_finance.info()
```

There are 617 films with at least a nonzero budget or nonzero revenue.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
plt.bar(['PG', 'R', 'G', 'PG-13', 'NR'],valid_finance['certification'].value_counts())
plt.title('Number of Films per Certification Category')
plt.xlabel('Certification')
plt.ylabel('Count')
```

# What is the average revenue per certification category?

```python
cert = valid_finance.groupby(['certification']).mean()
plt.bar(cert.index.values, cert['revenue'])
plt.title('Average Revenue per Certification')
plt.xlabel('Certification')
plt.ylabel('Average Revenue (in 100,000,000)');
```

# What is the average budget per certification category?

```python
plt.bar(cert.index.values, cert['budget'])
plt.title('Average Budget per Certification')
plt.xlabel('Certification')
plt.ylabel('Average Budget (in 10,000,000)');
```
