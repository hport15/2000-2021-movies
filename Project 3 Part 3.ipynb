```python
import pandas as pd
titleBasics = pd.read_csv('Data/title_basics.csv.gz')
titleBasics.head()
```

```python
# create a col w/list of genres
titleBasics['genres_split'] = titleBasics['genres'].str.split(",")
titleBasics
```

```python
# use .explode to separate genres_split into separate rows
explodedGenres = titleBasics.explode('genres_split')
```

```python
explodedGenres
```

```python
# save list of unique genres
uniqueGenres = sorted(explodedGenres['genres_split'].unique())
```

```python
# make title_genres table
title_genres = explodedGenres[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
# make genre map dict
genreInts = range(len(uniqueGenres))
genreMap = dict(zip(uniqueGenres, genreInts))
genreMap
```

```python
# add genreId col to title_genres
title_genres['genre_id'] = title_genres['genres_split'].map(genreMap)
title_genres
```

```python
# drop genres_split
title_genres.drop(columns=['genres_split'], inplace=True)
title_genres
```

```python
# make genres df
genres = pd.DataFrame({'genre_name':genreMap.keys(),
                      'genre_id': genreMap.values()})
genres.head()
```

```python
# save table in sql
title_genres.set_index('genre_id').to_sql('title_genres', engine, index=True,
                                         if_exists='replace')
```

```python
q = '''
SELECT *
FROM title_genres
LIMIT 5;'''
pd.read_sql(q, engine)
```

```python
# for genres, set df index and use index=True
genres.set_index('genre_id').to_sql('genres', engine, index=True)
```

```python
q = '''
SELECT *
FROM genres
LIMIT 5;'''
pd.read_sql(q, engine)
```

```python
engine.execute('ALTER TABLE genres ADD PRIMARY KEY (genre_id);')
```

```python
# title_basics
titleBasics.drop(columns=['originalTitle','isAdult', 'titleType', 'genres',
                  'genres_split'], inplace=True)
titleBasics
```

```python
# dtype schema for title_basics
from sqlalchemy.types import *
# calculate max str lengths
tconstLen = titleBasics['tconst'].fillna('').map(len).max()
titleLen = titleBasics['primaryTitle'].fillna('').map(len).max()
# make schema dict using sqlalchemy objects
tbSchema = {
    'tconst': String(tconstLen+1),
    'primaryTitle': Text(titleLen+1),
    'startYear': Float(),
    'endYear': Float(),
    'runtimeMinutes': Integer()
}
```

```python
# save title_basics table to sql
titleBasics.to_sql('title_basics', engine, dtype=tbSchema,
                   if_exists='replace',
                  index=False)
```

```python
# add tconst as primary key
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (tconst)')
```

```python
q = '''
SELECT *
FROM title_basics
LIMIT 5;'''
pd.read_sql(q, engine)
```

```python
tmdb = pd.read_csv('Data/tmdb_results_combined.csv.gz')
tmdb.head()
```

```python
tmdb.drop(index=tmdb.index[0], 
        axis=0, 
        inplace=True)
```

```python
tmdb.info()
```

```python
# add tmdb table to sql
# calculate max str lengths for object cols
imdbidLen = tmdb['imdb_id'].fillna('').map(len).max()
bdpLen = tmdb['backdrop_path'].fillna('').map(len).max()
btcLen = tmdb['belongs_to_collection'].fillna('').map(len).max()
genresLen = tmdb['genres'].fillna('').map(len).max()
hpLen = tmdb['homepage'].fillna('').map(len).max()
olLen = tmdb['original_language'].fillna('').map(len).max()
otLen = tmdb['original_title'].fillna('').map(len).max()
ovLen = tmdb['overview'].fillna('').map(len).max()
ppLen = tmdb['poster_path'].fillna('').map(len).max()
pcLen = tmdb['production_companies'].fillna('').map(len).max()
pcountLen = tmdb['production_countries'].fillna('').map(len).max()
rdLen = tmdb['release_date'].fillna('').map(len).max()
slLen = tmdb['spoken_languages'].fillna('').map(len).max()
statusLen = tmdb['status'].fillna('').map(len).max()
tlLen = tmdb['tagline'].fillna('').map(len).max()
titleLen = tmdb['title'].fillna('').map(len).max()
certLen = tmdb['certification'].fillna('').map(len).max()
# create dtype schema
tmdbSchema = {
    'imdb_id': String(imdbidLen+1),
    'adult': Float(),
    'backdrop_path': Text(bdpLen+1),
    'belongs_to_collection': Text(btcLen+1),
    'budget': Float(),
    'genres': Text(genresLen+1),
    'homepage': Text(hpLen+1),
    'id': Float(),
    'original_language': Text(olLen+1),
    'original_title': Text(otLen+1),
    'overview': Text(ovLen+1),
    'popularity': Float(),
    'poster_path': Text(ppLen+1),
    'production_companies': Text(pcLen+1),
    'production_countries': Text(pcountLen+1),
    'release_date': Text(rdLen+1),
    'revenue': Float(),
    'runtime': Float(),
    'spoken_languages': Text(slLen+1),
    'status': Text(statusLen+1),
    'tagline': Text(tlLen+1),
    'title': Text(titleLen+1),
    'video': Float(),
    'vote_average': Float(),
    'vote_count': Float(),
    'certification': Text(certLen+1)
}
```

```python
# save tmdb_data table in sql
tmdb.to_sql('tmdb_data', engine, dtype=tmdbSchema, if_exists='replace',
           index=False)
```

```python
# add imdb_id as primary key
engine.execute('ALTER TABLE tmdb_data ADD PRIMARY KEY (imdb_id);')
```

```python
q = '''
SELECT *
FROM tmdb_data
LIMIT 5;'''
pd.read_sql(q, engine)
```
