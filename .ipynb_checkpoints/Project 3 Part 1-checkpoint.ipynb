```python
# replace nulls
basics.replace({'\\N':np.nan}, inplace=True)
basics.head()
```

```python
# replace nulls
akas.replace({'\\N':np.nan}, inplace=True)
ratings.replace({'\\N':np.nan}, inplace=True)
```

```python
basics.info()
```

```python
# drop basics rows where runttimeMinutes and/or genre is nan
basics.dropna(subset=['runtimeMinutes', 'genres'], inplace=True)
```

```python
basics.info()
```

```python
# keep only basics df rows w/titleType Movie
basics = basics[basics['titleType']=='movie']
```

```python
# keep only basics df rows w/startYear between 2000-2021
basics = basics[(basics['startYear'] >= '2000') & (basics['startYear'] <= '2021')]
```

```python
# exclude documentaries in basics df
# find all rows w/genre documentary
isDoc = basics['genres'].str.contains('documentary', case=False)
basics = basics[~isDoc]
```

```python
akas.info()
```

```python
# region values
akas['region'].unique()
```

```python
# keep only us region for akas df
akas = akas[akas['region']=='US']
akas.info()
```

```python
# filter basics df to only include us movies by using akas df
usFilms = basics['tconst'].isin(akas['titleId'])
usFilms
```

```python
basics = basics[usFilms]
basics
```
